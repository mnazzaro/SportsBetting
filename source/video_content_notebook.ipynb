```python
import pandas as pd
import config

df = pd.read_csv(f'{config.TRAINING_DATA_PATH}/dataset.csv')
```

```python
""" This prints out the greatest fighters of all time according to ELO as of their last fight """

latest_fights = df.sort_values(by='date').groupby('fighter_red').tail(1)

latest_fights[['fighter_red', 'elo_red']].sort_values(by='elo_red', ascending=False).head(10)
```

```python
""" This prints out the greatest fighters of all time at the top of their career by max ELO """

max_elo_fighters = df.loc[df.groupby('fighter_red')['elo_red'].idxmax()]

max_elo_fighters[['fighter_red', 'elo_red']].sort_values(by='elo_red', ascending=False).head(25)
```
